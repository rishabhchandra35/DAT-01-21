```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
df= pd.read_csv('../data/titanic.csv')
```

```python
df.head()
```

```python
df.Sex.value_counts()
```

```python
df.Sex.value_counts().plot(kind="bar")
```

```python
df['Embarked'].plot(kind='bar')
#this doesn't work because there are no numerical values to plot. its categorical 
```

```python
df.groupby('Sex')['Survived'].mean().plot(kind='bar')
```

```python
df.groupby('Sex')['Survived'].mean().plot(kind='barh')
#same as above but horizontal
```

```python
df['Embarked'].value_counts().plot(kind='pie')
```

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')
```

```python
df['Embarked'].value_counts().plot(kind='pie'); 
#if you add semicolon, you won't get the messy <matplotlib.axes._subplots.AxesSubplot at 0x114f5ff90> thing
```

```python
df['Embarked'].value_counts().plot(kind='pie',colors=['blue','green','red']); 
```

```python
df['Age'].plot(kind='hist')
```

```python
df['Age'].plot(kind='hist',bins=20)
```

```python
df.hist(column='Age',by='Embarked')
# you can do this w histograms and boxplots only 
```

```python
df.boxplot(column='Age',by='Embarked')
```

```python
df.hist(column='Age',by='Embarked',layout=(1,3))
```

```python
df.hist(column='Age',by='Embarked',layout=(1,3), figsize=(12,4))
```

```python
df.hist(column='Age',by='Embarked',layout=(1,3), figsize=(12,4),sharey=True)
# this lets you scale it better,  You can also do sharex 
```

```python
df['Age'].plot(kind='kde')
# smoother version of histogram.  Though histogram shows you value counts and kde shows you % of all values 
```

```python
df.plot('Age','Fare',kind='scatter')
# scatter plot you don't specify the series 
```
