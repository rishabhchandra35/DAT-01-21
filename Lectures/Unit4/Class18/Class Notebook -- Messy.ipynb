```python
import pandas as pd
import numpy as np
```

```python
seeds   = pd.read_csv('../data/NCAA/MNCAATourneySeeds.csv')
results = pd.read_csv('../data/NCAA/MNCAATourneyCompactResults.csv')
```

```python
seeds.head()
```

```python
seeds['Seed'] = seeds['Seed'].str[1:].str[:2].astype(int)
```

```python
results = results[['Season', 'WTeamID', 'WScore', 'LTeamID', 'LScore']]
```

```python
results.head()
```

```python
results.columns = ['Season', 'T1TeamID', 'T1Score', 'T2TeamID', 'T2Score']
```

```python
results.head()
```

```python
results_swap = results[['Season', 'T2TeamID', 'T2Score', 'T1TeamID', 'T1Score']]
```

```python
results_swap.head()
```

```python
results_swap.columns = results.columns
results_swap.head()
```

```python
results.head()
```

```python
results['Result'] = 1
results_swap['Result'] = 0
```

```python
game_data = pd.concat([results, results_swap])
```

```python
game_data.head()
```

```python
seeds.head()
```

```python
game_data.head()
```

```python
seeds.head()
```

```python
game_data = game_data.merge(seeds, how='left', left_on=['Season', 'T1TeamID'], right_on=['Season', 'TeamID'])
```

```python
game_data.rename({'Seed': 'T1Seed'}, axis=1, inplace=True)
game_data.drop('TeamID', axis=1, inplace=True)
```

```python
game_data.head()
```

```python
game_data = game_data.merge(seeds, how='left', left_on=['Season', 'T2TeamID'], right_on=['Season', 'TeamID'])
```

```python
game_data.rename({'Seed': 'T2Seed'}, axis=1, inplace=True)
game_data.drop('TeamID', axis=1, inplace=True)
```

```python
game_data.head()
```

```python
game_data['SeedDiff'] = game_data['T1Seed'] - game_data['T2Seed']
game_data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

X = game_data[['SeedDiff']]
y = game_data['Result']
```

```python
X.shape
```

```python
y.shape
```

```python
rf.fit(X, y)
```

```python
rf.predict(X)
```

```python
rf.score(X, y)
```

```python
game_data['Prediction'] = np.where(rf.predict_proba(X)[:, 1] > 0.85, 1, 0)
```

```python
X.head(20)
```

For a classification tree:

 - continually find the best value to split on
 - split down until you have your individual leaves
 - prediction:  majority vote for all samples 
 - probabilities:  proportion of each outcome on the leaf
 
For 5 vs 11 seeds:

Total of 1000 samples on all of our trees

850 samples have value of 1
150 samples have value of 0

Prediction of ALL sample will be 1

predict_proba = .15 chance value is 0, 0.85 chance value is 1
